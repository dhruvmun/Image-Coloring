```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
from PIL import Image
import pickle
%matplotlib inline
```

```python
IMAGE_DIMENSION = 128
NO_OF_IMAGES = 10
IMG_PATH = 'dataset/beach'

input_image = np.zeros((NO_OF_IMAGES, IMAGE_DIMENSION, IMAGE_DIMENSION, 1))
output_image = np.zeros((NO_OF_IMAGES, IMAGE_DIMENSION, IMAGE_DIMENSION, 3))
```

```python
for i in range(NO_OF_IMAGES):
    IMAGE_PATH = IMG_PATH + str(i) + '.jpg'
    image = Image.open(IMAGE_PATH)
    image = image.resize((IMAGE_DIMENSION, IMAGE_DIMENSION))
    img_gray = image.convert('L')
    
    input_image[i,:,:,:] = np.asarray(img_gray, dtype=np.float128).reshape(IMAGE_DIMENSION, IMAGE_DIMENSION, 1)
    output_image[i,:,:,:] = np.asarray(image, dtype=np.float128)
```

```python
print(input_image.shape)
print(output_image.shape)
```

```python
fig = plt.figure(figsize=(15,5))

for i in range(10):
    ax = fig.add_subplot(2, 5, 1 + i, xticks=[], yticks=[])
    im = input_image[i].reshape(IMAGE_DIMENSION,IMAGE_DIMENSION)
    plt.imshow(im)
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))

for i in range(10):
    ax = fig.add_subplot(2, 5, 1 + i, xticks=[], yticks=[])
    im = output_image[i].astype((np.uint8))
    plt.imshow(im)
plt.show()
```

## Normalize

```python
mean = np.mean(input_image,axis=(0,1,2,3))
std = np.std(input_image,axis=(0,1,2,3))
print(mean)
print(std)
```

```python
input_image = (input_image-mean)/(std+1e-7)
```

```python
mean1 = np.mean(output_image,axis=(0,1,2))
std1 = np.std(output_image,axis=(0,1,2))
print(mean1)
print(std1)
```

```python
for i in range(3):
    output_image[:,:,:,i] = (output_image[:,:,:,i]-mean1[i])/(std1[i]+1e-7)
```

```python
print(input_image.shape)
print(output_image.shape)
```

```python
import cnn
```

```python
Cnn = cnn.CNN()
```

```python
J = Cnn.train_model(input_image, output_image, batch_size=10, iters=5)
```
