```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
from PIL import Image
%matplotlib inline
```

```python
IMG_WIDTH = 32
IMG_HEIGHT = 32
COLOR_CHANNELS = 3
```

```python
CONTENT_IMG_PATH = 'content1.jpg'
#content_image = scipy.misc.imread(CONTENT_IMG_PATH)
#content_image = np.reshape(content_image, ((1,IMG_HEIGHT,IMG_WIDTH,COLOR_CHANNELS)))
#imshow(content_image[0])

content_image = Image.open(CONTENT_IMG_PATH)
content_image
```

```python
output_image = content_image.resize((IMG_HEIGHT, IMG_WIDTH))
input_image = content_image.convert('L')
input_image = input_image.resize((IMG_HEIGHT, IMG_WIDTH))
```

```python
input_image = np.asarray(input_image, dtype='float')
input_image = np.expand_dims(input_image, axis=0)
input_image = input_image.reshape((1,32,32,1))
print(input_image.shape)
imshow(input_image[0])
```

```python
output_image = np.asarray(output_image,dtype='float')
output_image = np.expand_dims(output_image, axis=0)
output_image = output_image.reshape((1,32,32,3))
print(output_image.shape)
imshow(output_image[0])
```

```python
import cnn
```

```python
Cnn = cnn.CNN()
```

```python
J = Cnn.train_model(input_image, output_image, batch_size=1, iters=50)
```
